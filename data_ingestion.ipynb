```python
import os
import yfinance as yf
os.makedirs(os.path.join("data","raw"),exist_ok=True)
print("Fetching 5 years of S&P 500 data...")
df=yf.download("^GSPC",period="5y")
save_path=os.path.join("data","raw","sp500_raw.csv")
df.to_csv(save_path)
print(f"Success! Data saved to: {save_path}")
print(df.head())
```
